```python
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import QFileDialog, QLabel
from PyQt5.QtGui import QPixmap
import deep_learning
import chartcolumn


class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1113, 1037)
        font = QtGui.QFont()
        font.setFamily("Times New Roman")
        font.setPointSize(11)
        MainWindow.setFont(font)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.btnChoiceTrain = QtWidgets.QPushButton(self.centralwidget)
        self.btnChoiceTrain.setEnabled(True)
        self.btnChoiceTrain.setGeometry(QtCore.QRect(960, 10, 93, 31))
        self.btnChoiceTrain.setAutoDefault(False)
        self.btnChoiceTrain.setDefault(False)
        self.btnChoiceTrain.setObjectName("btnChoiceTrain")
        self.Train = QtWidgets.QLabel(self.centralwidget)
        self.Train.setGeometry(QtCore.QRect(50, 10, 71, 31))
        font = QtGui.QFont()
        font.setPointSize(12)
        self.Train.setFont(font)
        self.Train.setObjectName("Train")
        self.pathTrain = QtWidgets.QTextEdit(self.centralwidget)
        self.pathTrain.setGeometry(QtCore.QRect(150, 10, 751, 31))
        self.pathTrain.setObjectName("pathTrain")
        self.groupBox = QtWidgets.QGroupBox(self.centralwidget)
        self.groupBox.setGeometry(QtCore.QRect(10, 120, 981, 51))
        self.groupBox.setObjectName("groupBox")
        self.rbtRaF = QtWidgets.QRadioButton(self.groupBox)
        self.rbtRaF.setGeometry(QtCore.QRect(100, 20, 161, 20))
        self.rbtRaF.setObjectName("rbtRaF")
        self.rbtSvm = QtWidgets.QRadioButton(self.groupBox)
        self.rbtSvm.setGeometry(QtCore.QRect(350, 20, 95, 20))
        self.rbtSvm.setObjectName("rbtSvm")
        self.rbtDecTr = QtWidgets.QRadioButton(self.groupBox)
        self.rbtDecTr.setGeometry(QtCore.QRect(540, 20, 131, 20))
        self.rbtDecTr.setObjectName("rbtDecTr")
        self.rbtNaB = QtWidgets.QRadioButton(self.groupBox)
        self.rbtNaB.setGeometry(QtCore.QRect(770, 20, 131, 20))
        self.rbtNaB.setObjectName("rbtNaB")
        self.groupBox_2 = QtWidgets.QGroupBox(self.centralwidget)
        self.groupBox_2.setGeometry(QtCore.QRect(10, 170, 981, 51))
        self.groupBox_2.setObjectName("groupBox_2")
        self.rbtAE = QtWidgets.QRadioButton(self.groupBox_2)
        self.rbtAE.setGeometry(QtCore.QRect(160, 20, 95, 20))
        self.rbtAE.setObjectName("rbtAE")
        self.rbtVAE = QtWidgets.QRadioButton(self.groupBox_2)
        self.rbtVAE.setGeometry(QtCore.QRect(480, 20, 95, 20))
        self.rbtVAE.setObjectName("rbtVAE")
        self.rbtDAE = QtWidgets.QRadioButton(self.groupBox_2)
        self.rbtDAE.setGeometry(QtCore.QRect(780, 20, 95, 20))
        self.rbtDAE.setObjectName("rbtDAE")
        self.textBrowser = QtWidgets.QTextBrowser(self.centralwidget)
        self.textBrowser.setGeometry(QtCore.QRect(590, 270, 491, 301))
        self.textBrowser.setObjectName("textBrowser")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(80, 590, 401, 20))
        self.label.setObjectName("label")
        self.label_3 = QtWidgets.QLabel(self.centralwidget)
        self.label_3.setGeometry(QtCore.QRect(240, 990, 111, 20))
        self.label_3.setObjectName("label_3")
        self.labelRateX = QtWidgets.QLabel(self.centralwidget)
        self.labelRateX.setGeometry(QtCore.QRect(20, 270, 521, 311))
        self.labelRateX.setMinimumSize(QtCore.QSize(391, 0))
        font = QtGui.QFont()
        font.setBold(False)
        font.setWeight(50)
        self.labelRateX.setFont(font)
        self.labelRateX.setText("")
        self.labelRateX.setObjectName("labelRateX")
        self.labelUAC = QtWidgets.QLabel(self.centralwidget)
        self.labelUAC.setGeometry(QtCore.QRect(20, 610, 521, 381))
        self.labelUAC.setText("")
        self.labelUAC.setObjectName("labelUAC")
        self.btnStart = QtWidgets.QPushButton(self.centralwidget)
        self.btnStart.setEnabled(True)
        self.btnStart.setGeometry(QtCore.QRect(510, 230, 93, 31))
        self.btnStart.setAutoDefault(False)
        self.btnStart.setDefault(False)
        self.btnStart.setObjectName("btnStart")
        self.labelTime = QtWidgets.QLabel(self.centralwidget)
        self.labelTime.setGeometry(QtCore.QRect(590, 620, 491, 361))
        self.labelTime.setText("")
        self.labelTime.setObjectName("labelTime")
        self.label_4 = QtWidgets.QLabel(self.centralwidget)
        self.label_4.setGeometry(QtCore.QRect(780, 990, 201, 20))
        self.label_4.setObjectName("label_4")
        self.pathTest = QtWidgets.QTextEdit(self.centralwidget)
        self.pathTest.setGeometry(QtCore.QRect(150, 50, 751, 31))
        self.pathTest.setObjectName("pathTest")
        self.pathColumn = QtWidgets.QTextEdit(self.centralwidget)
        self.pathColumn.setGeometry(QtCore.QRect(150, 90, 751, 31))
        self.pathColumn.setObjectName("pathColumn")
        self.btnChoiceColumn = QtWidgets.QPushButton(self.centralwidget)
        self.btnChoiceColumn.setEnabled(True)
        self.btnChoiceColumn.setGeometry(QtCore.QRect(960, 90, 93, 31))
        self.btnChoiceColumn.setAutoDefault(False)
        self.btnChoiceColumn.setDefault(False)
        self.btnChoiceColumn.setObjectName("btnChoiceColumn")
        self.btnChoiceTest = QtWidgets.QPushButton(self.centralwidget)
        self.btnChoiceTest.setEnabled(True)
        self.btnChoiceTest.setGeometry(QtCore.QRect(960, 50, 93, 31))
        self.btnChoiceTest.setAutoDefault(False)
        self.btnChoiceTest.setDefault(False)
        self.btnChoiceTest.setObjectName("btnChoiceTest")
        self.Train_2 = QtWidgets.QLabel(self.centralwidget)
        self.Train_2.setGeometry(QtCore.QRect(20, 90, 111, 31))
        font = QtGui.QFont()
        font.setPointSize(12)
        self.Train_2.setFont(font)
        self.Train_2.setObjectName("Train_2")
        MainWindow.setCentralWidget(self.centralwidget)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)
        self.structure = 0
        self.algorithm = 0

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "Deeplearning"))
        self.btnChoiceTrain.setText(_translate("MainWindow", "Choice File"))
        self.Train.setText(_translate("MainWindow", "Data"))
        self.groupBox.setTitle(_translate("MainWindow", "Thuật toán"))
        self.rbtRaF.setText(_translate("MainWindow", "Random Forest"))
        self.rbtSvm.setText(_translate("MainWindow", "SVM"))
        self.rbtDecTr.setText(_translate("MainWindow", "Decision Tree"))
        self.rbtNaB.setText(_translate("MainWindow", "Naive Baves"))
        self.groupBox_2.setTitle(_translate("MainWindow", "Cấu trúc mạng"))
        self.rbtAE.setText(_translate("MainWindow", "AE"))
        self.rbtVAE.setText(_translate("MainWindow", "VAE"))
        self.rbtDAE.setText(_translate("MainWindow", "DAE"))
        self.label.setText(_translate("MainWindow", "Biểu đồ thống kê tỉ lệ mẫu tấn tông và không tấn công"))
        self.label_3.setText(_translate("MainWindow", "Biểu đồ UAC"))
        self.btnStart.setText(_translate("MainWindow", "Start train"))
        self.label_4.setText(_translate("MainWindow", "Biểu đồ thời gian phân loại"))
        self.btnChoiceColumn.setText(_translate("MainWindow", "Choice File"))
        self.btnChoiceTest.setText(_translate("MainWindow", "Choice File"))
        self.Train_2.setText(_translate("MainWindow", "Column name"))

    def take_event_click(self):
        self.btnChoiceTrain.clicked.connect(lambda: self.openFileTrain())
        self.btnChoiceTest.clicked.connect(lambda: self.openFileTest())
        self.btnChoiceColumn.clicked.connect(lambda: self.openFileColumn())
        self.btnStart.clicked.connect(lambda: self.deep_training(self.structure))
        self.rbtSvm.toggled.connect(lambda: self.take_event_radbtn(self.rbtSvm))
        self.rbtDecTr.toggled.connect(lambda: self.take_event_radbtn(self.rbtDecTr))
        self.rbtRaF.toggled.connect(lambda: self.take_event_radbtn(self.rbtRaF))
        self.rbtNaB.toggled.connect(lambda: self.take_event_radbtn(self.rbtNaB))
        self.rbtAE.toggled.connect(lambda: self.take_event_radbtn(self.rbtAE))
        self.rbtVAE.toggled.connect(lambda: self.take_event_radbtn(self.rbtVAE))
        self.rbtDAE.toggled.connect(lambda: self.take_event_radbtn(self.rbtDAE))

    def openFileTrain(self):
        options = QFileDialog.Options()
        options |= QFileDialog.DontUseNativeDialog
        filename, _ = QFileDialog.getOpenFileName(
            None, 'Open File', "/home/lehuy/DATA/Doan",
            'All_File (*)',
            None, QFileDialog.DontUseNativeDialog)
        self.pathTrain.setText(filename)

    def openFileTest(self):
        options = QFileDialog.Options()
        options |= QFileDialog.DontUseNativeDialog
        filename, _ = QFileDialog.getOpenFileName(
            None, 'Open File', "/home/lehuy/DATA/Doan",
            'All_File (*)',
            None, QFileDialog.DontUseNativeDialog)
        self.pathTest.setText(filename)

    def openFileColumn(self):
        options = QFileDialog.Options()
        options |= QFileDialog.DontUseNativeDialog
        filename, _ = QFileDialog.getOpenFileName(
            None, 'Open File', "/home/lehuy/DATA/Doan",
            'All_File (*)',
            None, QFileDialog.DontUseNativeDialog)
        self.pathColumn.setText(filename)

    def take_event_radbtn(self, radbtn):

        if radbtn.text() == "Random Forest" and radbtn.isChecked():
            self.algorithm = 1
        if radbtn.text() == "SVM" and radbtn.isChecked():
            self.algorithm = 2
        if radbtn.text() == "Decision Tree" and radbtn.isChecked():
            self.algorithm = 3
        if radbtn.text() == "Naive Baves" and radbtn.isChecked():
            self.algorithm = 4
        if radbtn.text() == "AE" and radbtn.isChecked():
            self.structure = 1
            # chartcolumn.take_content_train('1','2')
        if radbtn.text() == "VAE" and radbtn.isChecked():
            self.structure = 2
        if radbtn.text() == "DAE" and radbtn.isChecked():
            self.structure = 3

    def chart_rate_Train_Test(self, name):
        image = QPixmap('Results/Image/' + name + '/chart_column_' + name + '.png')

        self.labelRateX.setPixmap(image)

    def chart_line_AUC(self, name, dataname):
        if self.algorithm == 0:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_AUC_' + name + '.png')
        if self.algorithm == 1:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_AUC_' + name + '_RF' + '.png')
        if self.algorithm == 2:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_AUC_' + name + '_SVM' + '.png')
        if self.algorithm == 3:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_AUC_' + name + '_DT' + '.png')
        if self.algorithm == 4:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_AUC_' + name + '_NB' + '.png')

        self.labelUAC.setPixmap(image)

    def chart_line_Time(self, name, dataname):
        if self.algorithm == 0:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_Time_' + name + '.png')
        if self.algorithm == 1:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_Time_' + name + '_RF' + '.png')
        if self.algorithm == 2:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_Time_' + name + '_SVM' + '.png')
        if self.algorithm == 3:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_Time_' + name + '_DT' + '.png')
        if self.algorithm == 4:
            image = QPixmap('Results/Image/' + dataname + '/chart_line_Time_' + name + '_NB' + '.png')

        self.labelTime.setPixmap(image)

    def get_index_data_name(self):
        self.index_data = 0
        if 'unsw' in self.pathTrain.toPlainText():
            self.index_data = 0
        elif 'ctu13_8' in self.pathTrain.toPlainText():
            self.index_data = 1
        elif 'Ads' in self.pathTrain.toPlainText():
            self.index_data = 2
        elif 'Phishing' in self.pathTrain.toPlainText():
            self.index_data = 3
        elif 'Antivirus' in self.pathTrain.toPlainText():
            self.index_data = 4
        elif 'Spam' in self.pathTrain.toPlainText():
            self.index_data = 5
        return self.index_data

    def display_content_training(self):
        str = ''
        for i in chartcolumn.context:
            self.textBrowser.append(i)

    def deep_training(self, structure):
        data_index, input_dim, balance_rate, lr, block_size, num_epoch, hidden_layer, \
        step, X_train, X_test, Y_train, Y_test, \
        dt, label_dim, noise_factor, filter_size, data_shape, conf, dataname = deep_learning.make_init_deep_learning(
            self.get_index_data_name(), self.pathTrain.toPlainText(), self.pathTest.toPlainText(),
            self.pathColumn.toPlainText())

        self.chart_rate_Train_Test(dataname)

        if structure != 0:
            if structure == 1:
                deep_learning.call_AE_train(data_index, input_dim, balance_rate, lr, block_size, num_epoch,
                                            hidden_layer,
                                            step, X_train, X_test, Y_train, Y_test, dt, dataname, conf, self.algorithm)
                self.display_content_training()
                self.chart_line_AUC(dataname + '_AE', dataname)
                self.chart_line_Time(dataname + '_AE', dataname)
            if structure == 2:
                deep_learning.call_VAE_train(data_index, input_dim, balance_rate, lr, block_size, num_epoch,
                                             hidden_layer, step, X_train, X_test, Y_train, Y_test, dt, dataname, conf,
                                             self.algorithm)
                self.display_content_training()
                self.chart_line_AUC(dataname + '_VAE', dataname)
                self.chart_line_Time(dataname + '_VAE', dataname)
            if structure == 3:
                deep_learning.call_DAE_train(data_index, input_dim, balance_rate, lr, block_size, num_epoch,
                                             hidden_layer, step, X_train, X_test, Y_train, Y_test, dt, noise_factor,
                                             dataname, conf, self.algorithm)
                self.display_content_training()
                self.chart_line_AUC(dataname + '_DAE', dataname)
                self.chart_line_Time(dataname + '_DAE', dataname)

    def handle_data(self):
        self.take_event_click()


if __name__ == "__main__":
    import sys

    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
    MainWindow.show()
    ui.handle_data()
    sys.exit(app.exec_())

```
